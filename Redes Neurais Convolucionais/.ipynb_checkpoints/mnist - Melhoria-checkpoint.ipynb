```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras.layers.
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
position = 1
plt.imshow(X_train[position], cmap = 'gray')
plt.title('Classe '+str(y_train[position]))
```

```python
previsores_train = X_train.reshape(X_train.shape[0],
                                  28,28,1)
```

```python
previsores_train.shape
```

```python
print(previsores_train[0])
```

```python
previsores_teste = X_test.reshape(X_test.shape[0],
                                 28,28,1)
```

```python
previsores_teste.shape
```

```python
print(previsores_teste[0])
```

```python
print("",previsores_train.dtype,"\n",previsores_teste.dtype)
```

```python
previsores_train = previsores_train.astype('float32')
previsores_teste = previsores_teste.astype('float32')
```

```python
print("",previsores_train.dtype,"\n",previsores_teste.dtype)
```

```python
previsores_train /= 255
previsores_teste /= 255
```

```python
print(previsores_teste[0])
```

```python
classe_train = np_utils.to_categorical(y_train,10)
```

```python
classe_teste = np_utils.to_categorical(y_test,10)
```

```python
print("",classe_train,"\n")
```

```python
print(classe_teste)
```

```python
classificador = Sequential()
```

```python
classificador.add(Conv2D(32, (3,3), input_shape=(28,28,1), activation='relu'))#mudar o 32 pra 64
```

```python
classificador.add(MaxPooling2D(pool_size=(2,2)))
```

```python
classificador.add(Flatten())
```

```python
classificador.add(Dense(units=128, activation='relu'))
classificador.add(Dense(units=10, activation='softmax'))
```

```python
classificador.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
classificador.fit(previsores_train,classe_train,batch_size=128,epochs = 5, validation_data=(previsores_teste,classe_teste))
```

```python
resultado = classificador.evaluate(previsores_teste,classe_teste)
resultado
```
